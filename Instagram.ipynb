```python
import pandas as pd

# Load the dataset into a DataFrame
data = pd.read_csv("instagram_reach.csv")
```

```python
# Extract input features and target variables
X = data.drop(['Unnamed: 0', 'S.No', 'USERNAME', 'Caption', 'Likes','Hashtags'], axis=1)
y_likes = data['Likes']
y_time_since_posted = data['Time since posted']
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_likes_train, y_likes_test, y_time_train, y_time_test = train_test_split(X, y_likes, y_time_since_posted, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

# Create separate linear regression models for each target variable
likes_model = LinearRegression()
time_model = LinearRegression()

# Fit the models on the training data
likes_model.fit(X_train, y_likes_train)
time_model.fit(X_train, y_time_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions on the testing data
likes_pred = likes_model.predict(X_test)
time_pred = time_model.predict(X_test)

# Evaluate the model's performance
likes_mse = mean_squared_error(y_likes_test, likes_pred)
likes_r2 = r2_score(y_likes_test, likes_pred)

time_mse = mean_squared_error(y_time_test, time_pred)
time_r2 = r2_score(y_time_test, time_pred)

print("Likes - Mean Squared Error:", likes_mse)
print("Likes - R-squared:", likes_r2)

print("Time Since Posted - Mean Squared Error:", time_mse)
print("Time Since Posted - R-squared:", time_r2)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MultiLabelBinarizer

# Load the dataset into a DataFrame
data = pd.read_csv("instagram_reach.csv")

#Converting 'Time since posted' to numeric value
data['Time since posted'] = data['Time since posted'].apply(lambda x: x.split(' ')[0])
data['Time since posted'] = pd.to_numeric(data['Time since posted'])

# Extract input features and target variables
X = data.drop(['Unnamed: 0', 'S.No', 'USERNAME', 'Caption', 'Likes'], axis=1)
y_likes = data['Likes']
y_time_since_posted = data['Time since posted']

# Preprocess the 'Hashtags' column using one-hot encoding
mlb = MultiLabelBinarizer()
hashtags_encoded = pd.DataFrame(mlb.fit_transform(X['Hashtags'].str.split()), columns=mlb.classes_, index=X.index)
X_encoded = pd.concat([X.drop('Hashtags', axis=1), hashtags_encoded], axis=1)

# Split the data into training and testing sets
X_train, X_test, y_likes_train, y_likes_test, y_time_train, y_time_test = train_test_split(X_encoded, y_likes, y_time_since_posted, test_size=0.2, random_state=42)

# Create separate linear regression models for each target variable
likes_model = LinearRegression()
time_model = LinearRegression()

# Fit the models on the training data
likes_model.fit(X_train, y_likes_train)
time_model.fit(X_train, y_time_train)

# Make predictions on the testing data
likes_pred = likes_model.predict(X_test)
time_pred = time_model.predict(X_test)

# Evaluate the model's performance
likes_mse = mean_squared_error(y_likes_test, likes_pred)
likes_r2 = r2_score(y_likes_test, likes_pred)

time_mse = mean_squared_error(y_time_test, time_pred)
time_r2 = r2_score(y_time_test, time_pred)

print("Likes - Mean Squared Error:", likes_mse)
print("Likes - R-squared:", likes_r2)

print("Time Since Posted - Mean Squared Error:", time_mse)
print("Time Since Posted - R-squared:", time_r2)
```
